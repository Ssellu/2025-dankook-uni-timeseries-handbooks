# 시계열 데이터의 3대 요소: 추세, 계절성, 잔여

## 시계열 데이터 정의
일정한 시간 간격으로 수집된 시간에 따른 데이터
- **시간 의존성**: 데이터는 시간 순서에 따라 정렬되며, 이전 관측치가 이후 관측치에 영향을 줌
- **연속성**: 일정한 간격으로 지속적으로 측정 (예: 시간별, 일별, 월별, 분기별)
- **복합적 구조**: 여러 패턴의 조합으로 구성

### 시계열 데이터의 예
1. **기상 데이터**: 
   - 일별 온도, 강수량, 습도 등
   - 예: `2025-05-01: 25°C, 2025-05-02: 26°C`

2. **금융 데이터**:
   - 주식 가격, 환율, 거래량 등
   - 예: `2025-05-01: 100 USD, 2025-05-02: 102 USD`

3. **소비자 행동 데이터**:
   - 시간별 웹사이트 방문자 수, 구매 트렌드
   - 예: `2025-05-01 10:00: 500명 방문`

4. **경제 데이터**:
   - 월별 실업률, GDP 성장률
   - 예: `2025년 1월: 3.5% 실업률`

5. **의료 데이터**:
   - 환자의 시간별 심박수, 혈압
   - 예: `2025-05-01 08:00: 120/80 mmHg`

6. **교통 데이터**:
   - 시간대별 교통량, 지하철 승객 수
   - 예: `2025-05-01 07:00: 10,000명`

7. **영상 데이터**
   - 프레임 데이터: 초당 프레임 수(FPS, Frames Per Second)에 따라 시간 순서대로 정렬된 이미지 데이터
   - 분석 가능 요소: 객체의 움직임, 색상 변화, 패턴 등

## 시계열 분해와 시계열 데이터 구성 요소

### 1. 시계열 분해(Time Series Decomposition)란?
- 시계열 데이터를 분석 단계에서, 데이터의 패턴을 이해하기 위해 데이터의 주요 구성 요소인 추세(Trend), 계절성(Seasonality), 잔여(Residual)를 분리하는 과정   
![시계열 데이터 요소 분해](./images/01-00-01.png)

### 2. 시계열 데이터의 구성 요소 
1. **추세(Trend)**
   - 데이터의 장기적인 변화 패턴
   - 시간에 따라 증가하거나 감소하는 경향
   - 데이터의 전반적인 방향성
2. **계절성(Seasonality)**
   - 데이터의 주기적인 변동
   - 특정 시간 간격(예: 월별, 계절별)으로 반복되는 패턴  
    > Seasonality/Cyclicity/Periodicity   
    > 1. Seasonality
    >   - 일정한 시간 간격(예: 초, 하루, 일주일, 한 달, 1년 등)마다 반복적으로 나타나는 패턴
    >   - 반복 주기가 항상 일정하며, 주기가 사전에 알려져 있음
    >   - 주로 특성·자연·사회적 요인(예: 계절, 요일, 월 등)에 의해 발생
    >   - 예: 여름마다 아이스크림 판매가 증가, 매주 월요일마다 전력 사용량이 감소
    > 2. Cyclicity
    >   - 반복적으로 나타나지만, 반복 주기가 일정하지 않고 불규칙적인 패턴
    >   - 반복 주기가 고정되어 있지 않거나 명확하지 않음
    >   - 외부 환경 변화에 의해 발생
    >   - 예: 경기 침체와 호황이 수년에 한 번씩 반복되는 경제 사이클, 특정 산업의 장기적인 호황/불황
    > 3. Periodicity
    >   - 일정한 시간 간격(고정된 주기)로 반복
    >   - 데이터의 진동 등 일정 주기를 반복하는 패턴
    >   - 계절성(Seasonality)에 포함되는 요소
1. **잔여(Residual/Remainder)**
   - 추세와 계절성을 제거한 후 남은 데이터
   - 예측할 수 없는 불규칙한 변동

#### **'노이즈'와 '잔여'의 차이**
##### 노이즈(Noise)
- 데이터에 포함된 랜덤한 변동으로, 본질적으로 예측이 불가능한 요소
- 잔여의 일부로 포함될 수 있음
##### 잔여(Residual)
- 시계열 데이터에서 **추세(Trend)**와 **계절성(Seasonality)**을 제거한 후 남은 부분
- 데이터의 불규칙한 변동을 나타내며, 예측할 수 없는 요소를 포함
- 노이즈 외에도 모델링되지 않은 패턴이나 구조적인 변동이 포함
  
### 3. 시계열 분해의 종류
#### 1. 가법 모델(Additive Model):
$$
Y(t) = T(t) + S(t) + R(t)
$$
- (Y(t)): 관측값
- (T(t)): 추세
- (S(t)): 계절성
- (R(t)): 잔여
- 데이터의 **변동폭(Amplitude)**이 일정
- 계절성(Seasonality)이 데이터 값에 상관없이 일정한 크기로 나타남
- 데이터가 선형적인 패턴을 보일 때 적합
- 계절성의 크기가 데이터 값에 영향을 받지 않는 경우 사용
- 예:
    - 일별 온도 변화 (봄, 여름, 가을, 겨울의 온도 차이가 일정한 경우)
    - 월별 매출 데이터에서 계절적 변동이 일정한 경우

#### 2. 승법 모델(Multiplicative Model):
$$
Y(t) = T(t) \times S(t) \times R(t)
$$

- 데이터의 **변동폭(Amplitude)**이 시간에 따라 증가하거나 감소
- 계절성(Seasonality)이 데이터 값에 비례하여 변화
- 데이터가 비선형적인 패턴을 보일 때 적합
- 예:
    - 주식 시장 데이터 (가격이 높아질수록 변동폭도 커짐)
    - 월별 매출 데이터에서 매출이 증가함에 따라 계절적 변동도 커지는 경우
- 승법 모델은 로그 변환을 통해 가법 모델로 변환 가능
$$
(\log(Y(t)) = \log(T(t)) + \log(S(t)) + \log(R(t)))
$$

**MES 시스템 데이터는 가법 모델? 승법 모델?**
- 생산량, 품질 데이터:
    - 생산량 데이터는 계절적 요인(예: 수요 증가/감소)에 따라 변동할 수 있음
    - 변동폭이 일정하다면 가법 모델, 변동폭이 생산량에 비례한다면 승법 모델이 적합할 수 있음
- 장비 가동 데이터:
    - 장비 가동률은 일정한 패턴을 보일 가능성이 높아 가법 모델이 적합할 수 있음
- 공정 데이터:
    - 공정 데이터는 비선형적 변동을 보일 가능성이 있으므로 승법 모델이 적합할 수 있음
 
### 4. 활용 사례
- 예측 모델링: 추세와 계절성을 기반으로 미래 값을 예측
- 이상 탐지: 잔여(Residual)에서 비정상적인 변동 탐지
- 데이터 이해: 데이터의 구조적 패턴 분석

## 4. 전체 시각화 예시 코드

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# 1. 예제 데이터 생성
time = pd.date_range(start='2020-01-01', periods=100, freq='D')
data = 0.05 * np.arange(100) + np.random.normal(scale=1, size=100)
ts = pd.Series(data, index=time)

# 2. 시계열 분해
decomposition = seasonal_decompose(ts, model='additive', period=30)

# 3. 구성 요소 추출
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# 4. 시각화
plt.figure(figsize=(10, 8))
plt.subplot(411)
plt.plot(ts, label='Original Data')
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(trend, label='Trend', color='red')
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(seasonal, label='Seasonality', color='green')
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(residual, label='Residual', color='orange')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```
